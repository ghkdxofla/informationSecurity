```python
# Miller rabin algorithm
def test(n):
    k = 1
    q = 0
    if n % 2 == 0:
        #print("composite")
        return "composite"
    while ((n-1) % (2**k)) ==0 :
        
        q = (n-1) / (2**k)
        k = k + 1
        
    k = k - 1
    a = 2
    if a**q % n == 1:
        #print("inconclusive")
        return "inconclusive"
    for j in range (k - 1) :
        if a**((2**j)*q) % n == n-1:
            #print("inconclusive")
            return "inconclusive"
    #print("composite")
    return "composite"

def RSA_encrypt(n):
    m = 2
    e = 23
    c = m**e % n
    return c

def RSA_private_key(p, q):
    e = 23
    n = p * q
    pi_n = (p - 1) * (q - 1)
    for d in range(n+1):
        if((e * d) % pi_n) == 1:
            return d

def RSA_decrypt(c, d, n):
    m = (c**d) % n
    return m


def CRT_56153(c, n):
    m1 = 233
    m2 = 241
    
    # Compute c^d mod m1, c^d mod m2
    # d를 구하기 위해 아래의 private key를 구할 시
    # "Square and Multiply” Algorithm을 사용하였습니다
    
    mod_233 = []
    mod_241 = []
    
    for i in range (15):
        mod_233.append((c**(2**i)) % 233)
        mod_241.append((c**(2**i)) % 241)

    # private key : 19367 = 16384 + 2048 + 512 + 256 + 128 + 32 + 4 + 2 + 1
    
    a1 = (mod_233[14]*mod_233[11]*mod_233[9]*mod_233[8]*mod_233[7]*
          mod_233[5]*mod_233[2]*mod_233[1]*mod_233[0])
    a2 = (mod_241[14]*mod_241[11]*mod_241[9]*mod_241[8]*mod_241[7]*
          mod_241[5]*mod_241[2]*mod_241[1]*mod_241[0])
    
    a1 = a1 % m1
    a2 = a2 % m2
        
    # 역원은 계산기를 사용하여 구하였습니다(사진 첨부)
    # 241^-1 mod 233 = 204
    # 233^-1 mod 241 = 30
    
    c1 = 241 * 204
    c2 = 233 * 30
    
    # c1, c2
    k1 = a1 * c1
    k2 = a2 * c2
    
    # find the message, m
    m = (k1 + k2) % 56153
    return m
    
    
    
    

# 소수 판별
print("Is {0} a prime number? : {1}".format(233, test(233)))
print("Is {0} a prime number? : {1}".format(234, test(234)))
print("Is {0} a prime number? : {1}".format(235, test(235)))
print("Is {0} a prime number? : {1}".format(236, test(236)))
print("Is {0} a prime number? : {1}".format(237, test(237)))
print("Is {0} a prime number? : {1}".format(238, test(238)))
print("Is {0} a prime number? : {1}".format(239, test(239)))
print("Is {0} a prime number? : {1}".format(240, test(240)))
print("Is {0} a prime number? : {1}".format(241, test(241)))
# 233, 239, 241 소수

# Encrypt the message M
c = RSA_encrypt(56153)
print("ciphertext : {0}".format(c))

# Private key using prime 233, 239, 241
d1 = RSA_private_key(233, 239)
d2 = RSA_private_key(239, 241)
d3 = RSA_private_key(233, 241)
print("private key : {0}".format(d1))
print("private key : {0}".format(d2))
print("private key : {0}".format(d3))

# Decrypt the ciphertext C
print("message : {0}".format(RSA_decrypt(c, d1, 56153)))
print("message : {0}".format(RSA_decrypt(c, d2, 56153)))
print("message : {0}".format(RSA_decrypt(c, d3, 56153)))

# Answer is p = 233, p = 241
# 이를 통해 위의 private key, 19367임을 알 수 있다

# CRT (56153 = 233 * 241)
print("message : {0}".format(CRT_56153(c, 56153)))

# CRT로 나온 message도 같으므로 답임을 알 수 있음
        
```
